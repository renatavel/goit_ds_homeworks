```python
import matplotlib.pyplot as plt
import seaborn as sns

levels = df['Рівень англійської'].value_counts() 
percentages = levels / levels.sum() * 100  

colors = sns.color_palette("pastel", len(levels))

plt.pie(
    percentages,
    labels=levels.index,
    autopct="%.1f%%",
    pctdistance=0.85,
    labeldistance=1.1,  
    textprops={'fontsize': 12},
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.0},
    colors=colors
)

plt.title("Level of Knowledge of English language", fontsize=19, pad=20, fontweight='bold')
plt.show()
```

```python
df_salary = df.groupby('Посада').agg({'Зарплата на місяць': ['min', 'max']})
df_salary
```

```python
df_salary = df_salary.reset_index()
```

```python
import matplotlib.pyplot as plt

positions = df_salary['Посада']
min_salaries = df_salary[('Зарплата на місяць', 'min')]
max_salaries = df_salary[('Зарплата на місяць', 'max')]

plt.bar(positions, min_salaries, width=0.4, label='Min Salary', color='red', align='center')
plt.bar(positions, max_salaries, width=0.4, label='Max Salary', color='blue', align='edge')

plt.xlabel('Job Position', fontsize="large")
plt.ylabel('Salary', fontsize="large")
plt.title('Min and Max Salaries Comparison by Job Position', fontweight = "book")
plt.xticks(rotation=45, ha='right')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def fill_avg_salary(row):
    return (row['Зарплата на місяць'].min() + row['Зарплата на місяць'].max())/2

df_salary['avg'] = df_salary.apply(lambda row: fill_avg_salary(row), axis=1)

df_salary
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.barh(df_salary['Посада'], df_salary['avg'], color='lightgreen')
plt.xlabel('Average Salary', fontsize= 17, labelpad=15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=12)
plt.title('Average Salary for Job Position', pad=20, fontsize=20)
plt.tight_layout()
plt.show()
```
